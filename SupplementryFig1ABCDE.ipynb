```python
import scanpy as sc
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import umap
#import matplotlib.colors as mcolors
#sc.settings.verbosity = 2
from adjustText import adjust_text
```

```python
# output figure parameter
sc.settings.set_figure_params(dpi=100, fontsize=12, dpi_save=400, figsize=(5,5))
```

```python
#cellenics/Trailmaker processed file
adata = sc.read('/home/zia/Downloads/Lc_data_final/LCRMCV1.h5ad')
```

```python
#legend outside the plot
def one_col_lgd(umap):
    legend = umap.legend(bbox_to_anchor=[1.00, 0.5],
                         loc='center left', ncol=1, prop={'size': 6})
    
    # Set legend frame alpha to make it transparent
    legend.get_frame().set_alpha(0.0)
    
    for handle in legend.legend_handles:
        handle.set_sizes([25.0])
    return legend
```

```python
#SupFig1A
CT_umap = sc.pl.umap(adata, color=['seurat_clusters'],
                    show=False, palette=sns.husl_palette(n_colors=26, l=0.5),
                    legend_fontsize=8, frameon=True, title='')
lgd = one_col_lgd(CT_umap)
fig = CT_umap.get_figure()
fig.patch.set_alpha(0.0) 
fig.set_size_inches(5, 5)
fig.savefig('/home/zia/Downloads/lc_final/seurat_clusters.pdf', dpi=600, bbox_extra_artists=(lgd,), bbox_inches='tight', transparent=True)
```

```python
#SupFig1B
cluster_counts = adata.obs['seurat_clusters'].value_counts()
sns.set(style="white")
plt.figure(figsize=(18, 8))
ax = sns.barplot(x=cluster_counts.index, y=cluster_counts.values, edgecolor='black', linewidth=0.5)
ax.tick_params(axis='x', labelsize=10, rotation=90)
ax.tick_params(axis='y', labelsize=10)
ax.set_xlabel('Cell Cluster', fontsize=12)
ax.set_ylabel('Number of Cells', fontsize=12)
ax.set_title('Number of Cells per Cluster', fontsize=12)
plt.tight_layout()
fig = plt.gcf()
fig.set_size_inches(6, 5)
plt.savefig('/home/zia/Downloads/lc_final/barplot.pdf', dpi=600, transparent=True, bbox_inches='tight')
plt.show()
```
